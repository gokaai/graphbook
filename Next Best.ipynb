```python
def get_info_button(tooltip_custom='', margin_custom='1px 3px 0px 10px'):
    b = Button(tooltip=tooltip_custom,
               icon='info',
               style={'button_color': 'transparent'},
               layout=Layout(width='27px',
                             height='30px',
                             margin=margin_custom,
                             border='solid 1px #0082ba'))
    b.add_class('info_button')
    return b

def get_file_refresh_button(margin_custom='1px 3px 0px 10px'):
    b = Button(tooltip='Refresh File List',
               icon='retweet',
               style={'button_color': 'transparent'},
               layout=Layout(width='40px',
                             margin=margin_custom,
                             border='solid 1px #0082ba'))
    b.add_class('info_button')
    return b


# display(Button(icon='info',button_style='info'))
# get_info_button()
# get_file_refresh_button()
```

```python
distinct_column_values = [
    'Distinct Value #1', 'Distinct Value #2', 'Distinct Value #3',
    'Distinct Value #4'
]


def get_assign_priority_table(combobox):
    cp_nb_priority_title = HTML('<strong>Assign 100% Reach for {}:'.format(
        cp_nb_hcp_priority_column_dropdown.value))
    cp_nb_priority_header = HBox([
        HTML('<p style="background-color:#ebebeb;">{}'.format(
            cp_nb_hcp_priority_column_dropdown.value),
             layout=Layout(width='15%')),
        HTML('<p style="background-color:#ebebeb;">100% Reach? (Y/N)',
             layout=Layout(width='15%'))
    ])
    #     cp_nb_assign_reach_box.children += tuple([cp_nb_priority_header])
    options_hbox = []
    for column_value in distinct_column_values:
        row_hbox = HBox([
            HTML(value=column_value,
                 layout=Layout(width='15%',
                               border='solid 1px',
                               padding='0px 0px 0px 10px')),
            Dropdown(options=['Yes', 'No'],
                     value=None,
                     layout=Layout(width='15%', padding='2px 2px 0px 2px'))
        ])
        options_hbox.append(row_hbox)

    cp_nb_assign_reach_box.children = [
        cp_nb_priority_title, cp_nb_priority_header
    ] + options_hbox


call_planning_master_button = Button(description='Call Planning',
                                     icon='phone',
                                     button_style='warning',
                                     style=ButtonStyle(button_color='#99cde3'),
                                     layout=Layout(width='22.5%',
                                                   margin='0px 0px 0px 0px'))


def show_next_best(obj):
    call_planning_master_button.style = ButtonStyle(button_color='#0082ba')
    display(cp_nb_master_tab)


call_planning_master_button.on_click(show_next_best)

cp_nb_master_tab = Tab([Text()], layout=Layout(width='100%'))
cp_nb_master_tab.set_title(0, 'Next Best Design/Refresh')
cp_nb_select_universe = Dropdown(options=[
    'Sample Universe File #1', 'Sample Universe File #2',
    'Sample Universe File #3'
],
                                 value=None,
                                 layout=Layout(width='22.5%'))
cp_nb_select_universe_description = HTML('<strong> Select Eligible Master:',
                                         layout=Layout(width='21%'))
cp_nb_select_universe_box = HBox([
    cp_nb_select_universe_description, cp_nb_select_universe,
    get_info_button(tooltip_custom='')
])

cp_nb_select_call_plan = Dropdown(options=[
    'Sample Call Plan File #1', 'Sample Call Plan File #2',
    'Sample Call Plan File #3'
],
                                  value=None,
                                  layout=Layout(width='22.5%'))
cp_nb_select_call_plan_description = HTML('<strong> Select Call Plan:',
                                          layout=Layout(width='21%'))
cp_nb_select_call_plan_box = HBox([
    cp_nb_select_call_plan_description, cp_nb_select_call_plan,
    get_info_button(tooltip_custom='')
])

# cp_nb_proritization_description = HTML('<strong> Set Prioritization:',
#                                        layout=Layout(width='10%'))
cp_nb_hcp_priority_column_description = HTML('HCP Priority Column Name:',
                                             layout=Layout(
                                                 width='21%',
                                                 margin='20px 2px 2px 2px'))
cp_nb_hcp_priority_column_text = Text(placeholder='Enter Output Column Name',
                                      layout=Layout(width='22.5%',
                                                    margin='20px 2px 2px 2px'))
cp_nb_hcp_priority_column_box = HBox([
    cp_nb_hcp_priority_column_description, cp_nb_hcp_priority_column_text,
    get_info_button(margin_custom='19px 3px 0px 10px')
])

cp_nb_hcp_priority_case_add_parameter = Button(
    icon='plus',
    layout=Layout(width='40px', margin='2px 3px 0px 2px'),
    style=ButtonStyle(button_color='#6c9b00'),
    button_style='success')

cp_nb_hcp_priority_case_delete_parameter = Button(
    icon='trash',
    layout=Layout(width='40px', margin='2px 3px 0px 10px'),
    style=ButtonStyle(button_color='#c10000'),
    button_style='success')

cp_nb_hcp_priority_case_description = HTML('Add/Remove Condition:',
                                           layout=Layout(width='21%'))
cp_nb_hcp_priority_case_row = HBox([
    cp_nb_hcp_priority_case_description, cp_nb_hcp_priority_case_add_parameter,
    cp_nb_hcp_priority_case_delete_parameter,
    get_info_button(
        'If-else statement that can be used to create ‘HCP Priority’.\
\nPlease note, similar to how is it done in data transformation tab,\
in case the HCP satisfies two if-else conditions, priority would be \
given to the last statement.')
])

cp_nb_if_description = HTML('<i>IF Clause')
cp_nb_if_text = Textarea(placeholder='Enter IF case expression',
                         layout=Layout(width='30%', margin='2px 2px 2px 5px'))
cp_nb_then_description = HTML('<i>Then',
                              layout=Layout(margin='2px 2px 2px 20px'))
cp_nb_then_text = Textarea(placeholder='Enter expression',
                           layout=Layout(width='30%',
                                         margin='2px 2px 2px 5px'))

cp_nb_else_description = HTML('<i>Else')
cp_nb_else_text = Textarea(placeholder='Enter Else case expression',
                           layout=Layout(width='30%',
                                         margin='2px 2px 2px 5px'))

cp_nb_priority_column_generate_button = Button(
    description='Generate Priority Column',
    style=ButtonStyle(button_color='#6c9b00'),
    button_style='success',
    layout=Layout(width='13%', margin='13px 2px 2px 55px'))
cp_nb_else_box = HBox([
    HTML('', layout=Layout(width='18.5%')), cp_nb_else_description,
    cp_nb_else_text, cp_nb_priority_column_generate_button
])

cp_nb_if_then_row = HBox([
    HTML('', layout=Layout(width='14%')),
    HTML('1.&emsp;'), cp_nb_if_description, cp_nb_if_text,
    cp_nb_then_description, cp_nb_then_text
])

cp_nb_if_then_box = VBox([cp_nb_if_then_row])

cp_nb_hcp_priority_box = VBox([
    cp_nb_hcp_priority_column_box,
    HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
    cp_nb_hcp_priority_case_row,
    HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')), cp_nb_if_then_box,
    HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')), cp_nb_else_box
])


def cp_nb_add_params(button):
    cp_nb_if_then_box.children += tuple([
        HBox([
            HTML('', layout=Layout(width='14%')),
            HTML('{}.&emsp;'.format(len(cp_nb_if_then_box.children) + 1)),
            cp_nb_if_description, cp_nb_if_text, cp_nb_then_description,
            cp_nb_then_text
        ])
    ])


def cp_nb_delete_params(button):
    if len(cp_nb_if_then_box.children) > 1:
        cp_nb_if_then_box.children = cp_nb_if_then_box.children[:-1]


cp_nb_parameters_description = HTML('<strong> Select Inputs:',
                                    layout=Layout(width='9%'))

cp_nb_universe_abbot_id_description = HTML(
    'Abbot Customer ID Column in Eligible Master:', layout=Layout(width='21%'))
cp_nb_universe_abbot_id_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_universe_abbot_id_box = HBox([
    cp_nb_universe_abbot_id_description, cp_nb_universe_abbot_id_dropdown,
    get_info_button(
        'Select the Abbot ID Column name that should be used from the Eligible Master:'
    )
])

cp_nb_call_plan_abbot_id_description = HTML(
    'Abbot Customer ID Column in Call Plan:', layout=Layout(width='21%'))
cp_nb_call_plan_abbot_id_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_call_plan_abbot_id_box = HBox([
    cp_nb_call_plan_abbot_id_description, cp_nb_call_plan_abbot_id_dropdown,
    get_info_button(
        'Select the Abbot ID Column name that should be used from the Call Plan Master file.'
    )
])

cp_nb_universe_territory_description = HTML(
    'Territory Column in Eligible Master:', layout=Layout(width='21%'))
cp_nb_universe_territory_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_universe_territory_box = HBox([
    cp_nb_universe_territory_description, cp_nb_universe_territory_dropdown,
    get_info_button(
        'Select the Territory Column name that should be used from the Eligibility Master file.'
    )
])

cp_nb_call_plan_territory_description = HTML('Territory Column in Call Plan:',
                                             layout=Layout(width='21%'))
cp_nb_call_plan_territory_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_call_plan_territory_box = HBox([
    cp_nb_call_plan_territory_description, cp_nb_call_plan_territory_dropdown,
    get_info_button(
        'Select the Territory Column name that should be used from the Call Plan Master file.'
    )
])

cp_nb_territory_rank_description = HTML('Territory Level Rank Column:',
                                        layout=Layout(width='21%'))
cp_nb_territory_rank_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_territory_rank_box = HBox([
    cp_nb_territory_rank_description, cp_nb_territory_rank_dropdown,
    get_info_button(
        'Select the Territory Level Rank Column. \nThis Column is used to recalculate the rank; \
records are sorted using this column and new rank is calculated name that should be used from the Call Plan Master file.'
    )
])

cp_nb_hcp_priority_column_description = HTML('HCP Priority Column Name:',
                                             layout=Layout(width='21%'))
cp_nb_hcp_priority_column_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_hcp_priority_column_box = HBox([
    cp_nb_hcp_priority_column_description, cp_nb_hcp_priority_column_dropdown,
    get_info_button(
        'Select the Territory Column name that should be used from the Call Plan Master file.'
    )
])

cp_nb_hcp_priority_column_dropdown.observe(get_assign_priority_table,
                                           names='value')

cp_nb_eligibility_column_description = HTML(
    'Eligibility Flag Column in Eligibility Master:',
    layout=Layout(width='21%'))
cp_nb_eligibility_column_dropdown = Dropdown(
    options=['ColumnName1', 'ColumnName2', 'ColumnName3'],
    value=None,
    layout=Layout(width='22.5%'))
cp_nb_eligibility_column_box = HBox([
    cp_nb_eligibility_column_description, cp_nb_eligibility_column_dropdown,
    get_info_button(
        'Select the Eligibility Flag Column name for next best, present in Eligibility Master.'
    )
])

cp_nb_parameters_box = VBox([
    cp_nb_parameters_description, cp_nb_universe_abbot_id_box,
    cp_nb_call_plan_abbot_id_box, cp_nb_universe_territory_box,
    cp_nb_call_plan_territory_box, cp_nb_territory_rank_box,
    cp_nb_hcp_priority_column_box, cp_nb_eligibility_column_box
])

cp_nb_final_column_names_description = HTML(
    '<strong> Assign Output Column Name:', layout=Layout(width='15%'))
cp_nb_final_frequency_column_description = HTML(
    'Name of the Final Frequency Column:', layout=Layout(width='21%'))
cp_nb_final_frequency_column_text = Text(placeholder='Calls',
                                         layout=Layout(width='22.5%'))
cp_nb_final_frequency_column_box = HBox([
    cp_nb_final_frequency_column_description,
    cp_nb_final_frequency_column_text,
    get_info_button(tooltip_custom='')
])

cp_nb_final_column_names_box = VBox(
    [cp_nb_final_column_names_description, cp_nb_final_frequency_column_box])

cp_nb_numeric_parameters_description = HTML(
    '<strong> Next Best Length Limits:', layout=Layout(width='15%'))
cp_nb_minimum_number_description = HTML('Minimum Number of Targets/Territory:',
                                        layout=Layout(width='21%'))
# cp_nb_minimum_number_intslider = BoundedIntText(value=60,
#                                            min=0,
#                                            max=120,
#                                            step=1,
#                                            layout=Layout(width='27.5%'))
cp_nb_minimum_number_intslider = BoundedIntText(value=60,
                                                min=0,
                                                max=120,
                                                step=1,
                                                layout=Layout(width='22.5%'))

cp_nb_minimum_number_box = HBox([
    cp_nb_minimum_number_description, cp_nb_minimum_number_intslider,
    get_info_button(tooltip_custom='')
])

cp_nb_maximum_number_description = HTML('Maximum Number of Targets/Territory:',
                                        layout=Layout(width='21%'))
cp_nb_maximum_number_intslider = BoundedIntText(value=80,
                                                min=0,
                                                max=120,
                                                step=1,
                                                layout=Layout(width='22.5%'))

cp_nb_maximum_number_box = HBox([
    cp_nb_maximum_number_description, cp_nb_maximum_number_intslider,
    get_info_button(tooltip_custom='')
])

cp_nb_numeric_parameters_box = VBox([
    cp_nb_numeric_parameters_description, cp_nb_minimum_number_box,
    cp_nb_maximum_number_box
])

cp_nb_final_column_names_box = VBox(
    [cp_nb_final_column_names_description, cp_nb_final_frequency_column_box])

cp_nb_assign_reach_box = VBox([])

cp_nb_final_generate_button = Button(description='Generate Next Best',
                                     style=ButtonStyle(button_color='#6c9b00'),
                                     button_style='success')

cp_nb_priority_radiobutton_description = HTML(
    '<strong>Select Option for Priority Column:')
cp_nb_priority_radiobutton = RadioButtons(options=[
    'Use Existing HCP Priority Column in Eligible Master',
    'Create a New HCP Priority Column in Eligible Master'
],
                                          layout=Layout(width='50%'))


def add_or_remove_priority_box(radiobutton):
#     print(cp_nb_priority_radiobutton.value.lower())
    if 'new' in cp_nb_priority_radiobutton.value.lower():
        cp_nb_hcp_priority_column_box.children = [
            cp_nb_hcp_priority_column_description,
            HTML(
                'Column Name will be picked from the name of the column generated above'
            )
        ]

        cp_nb_master_tab.children = [
            VBox([
                cp_nb_select_universe_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_select_call_plan_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_priority_radiobutton_box, cp_nb_hcp_priority_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_parameters_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_final_column_names_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_numeric_parameters_box, cp_nb_assign_reach_box,
                HTML('<hr>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_final_generate_button
            ])
        ]
    else:
        cp_nb_hcp_priority_column_box.children = [
            cp_nb_hcp_priority_column_description,
            cp_nb_hcp_priority_column_text,
            get_info_button(margin_custom='19px 3px 0px 10px')
        ]

        cp_nb_master_tab.children = [
            VBox([
                cp_nb_select_universe_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_select_call_plan_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_priority_radiobutton_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_parameters_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_final_column_names_box,
                HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_numeric_parameters_box, cp_nb_assign_reach_box,
                HTML('<hr>', layout=Layout(margin='0px 0px 0px 0px')),
                cp_nb_final_generate_button
            ])
        ]


cp_nb_priority_radiobutton_box = VBox(
    [cp_nb_priority_radiobutton_description, cp_nb_priority_radiobutton])

cp_nb_priority_radiobutton.observe(add_or_remove_priority_box, names='value')

cp_nb_master_tab.children = [
    VBox([
        cp_nb_select_universe_box,
        HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
        cp_nb_select_call_plan_box,
        HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
        cp_nb_priority_radiobutton_box,
        HTML('<br>',
             layout=Layout(margin='0px 0px 0px 0px')), cp_nb_parameters_box,
        HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
        cp_nb_final_column_names_box,
        HTML('<br>', layout=Layout(margin='0px 0px 0px 0px')),
        cp_nb_numeric_parameters_box, cp_nb_assign_reach_box,
        HTML('<hr>', layout=Layout(margin='0px 0px 0px 0px')),
        cp_nb_final_generate_button
    ])
]

cp_nb_hcp_priority_case_add_parameter.on_click(cp_nb_add_params)
cp_nb_hcp_priority_case_delete_parameter.on_click(cp_nb_delete_params)
# get_assign_priority_table(None)
```
